```python
amazon['actual_price']=amazon['actual_price'].str.replace('₹', '')
amazon['actual_price']=amazon['actual_price'].str.replace(',', '')
amazon['rating']=amazon['rating'].str.replace(',', '')
amazon['rating_count']=amazon['rating_count'].str.replace(',', '')
amazon['discounted_price']=amazon['discounted_price'].str.replace('₹', '')
amazon['discounted_price']=amazon['discounted_price'].str.replace(',', '')
```

```python
amazon['actual_price']=amazon['actual_price'].astype(float)
amazon['discounted_price']=amazon['discounted_price'].astype(float)
amazon['rating_count']=amazon['rating_count'].astype(float)
total_price=amazon['actual_price'].sum()
```

```python
import seaborn as sns
```

```python
plt.figure()
sns.histplot(amazon['actual_price'], bins=15)
plt.title("Distribution of Product Prices")
plt.show()
```

```python
top10 = amazon.sort_values(by='actual_price', ascending=False).head(10)

plt.figure()
sns.barplot(x='actual_price', y='product_name', data=top10)
plt.title("Top 10 Most Expensive Products")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='actual_price', y='rating', data=amazon)
plt.title("Price vs Rating")
plt.show()
```
