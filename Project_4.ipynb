```python
sns.countplot(x='type',data=netflix)
plt.title('Movie vs TV shows')
plt.show()
```

```python
top_country=netflix['country'].value_counts().head(5)
top_country
```

```python
top_country=netflix['country'].value_counts().head(5)
sns.barplot(x=top_country.values,y=top_country.index)
plt.title("Top 5 Country")
plt.show()
```

```python
sns.countplot(y='rating',data=netflix,order=netflix['rating'].value_counts().index)
plt.title('content rating')
plt.show()
```

```python
netflix.columns
```

```python
netflix['year_added'] = pd.to_datetime(netflix['date_added'], errors='coerce').dt.year

sns.countplot(x="year_added", data=netflix)
plt.xticks(rotation=90)
plt.title("Content Added Per Year")
plt.show()
```

```python
top_years = netflix['release_year'].value_counts().head(10)

sns.barplot(x=top_years.index, y=top_years.values)

plt.xlabel("Release Year")
plt.ylabel("Count")
plt.title("Top 10 Release Years")

plt.show()
```

```python
top_years = movies['release_year'].value_counts().head(10)

sns.barplot(x=top_years.index, y=top_years.values)
plt.xticks(rotation=45)
plt.title("Top 10 Movie Release Years")
plt.show()
```
