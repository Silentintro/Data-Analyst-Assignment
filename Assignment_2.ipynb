### 3.What are missing values in Pandas and how are they represented? Create a DataFrame with missing values and      write code to:
###   •Detect missing values
###   •Replace them with the column mean Explain each step.
    Ans: Missing values represent absent data, represented by: NaN and None.

```python
import pandas as pd
dt = {
    "Name": ["Ajay", "Anjali", "Harsh", "Priya"],
    "Math": [82, np.nan, 70, 95],
    "Science": [88, 65, np.nan, 90]
}
data = pd.DataFrame(dt)
print(data)
```

```python
data["Math"] = data["Math"].fillna(data["Math"].mean())
data["Science"] = data["Science"].fillna(data["Science"].mean())
data
```

### 4.Explain boolean indexing in NumPy or Pandas.            Create a DataFrame with at least 5 columns and filter rows based on two conditions using logical operators. Explain how filtering works internally.
    Ans: Boolean indexing is used for filter data in pandas to specified condition 

```python
data={'Name': ['harsh', 'hemant', 'ayush', 'mohit', 'aman'],
    'Age': [21, 22, 22, 21, 23],
    'Salary': [50000, 80000, 120000, 90000, 95000],
    'Dept': ['EE', 'CSE', 'CSE', 'CSE', 'ME']
}       
df=pd.DataFrame(data)
filter_data=df[(df['Salary']>60000) & (df['Dept']=='CSE')]
print(f"The Filter data is:\n{filter_data}")
```

### 5.What is the purpose of the groupby() function in Pandas? Create a DataFrame with categorical data (e.g., department & salary) and calculate the average salary per department using groupby(). Explain the output.
    Ans: Groupby() groups data based on categorical columns and applies aggregation functions.

```python
data = pd.DataFrame({
    'Department': ['accounts', 'Teaching', 'drivers', 'Transport', 'adminstrator'],
    'Salary': [750000, 40000, 20000, 80000, 90000]
})
avg_salary = data.groupby('Department')['Salary'].mean()
print(f"The average of salary:{avg_salary}")
```
