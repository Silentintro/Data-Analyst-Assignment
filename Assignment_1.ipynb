### 2.Write a program that takes a list of elements and creates a dictionary showing the frequency of each element.

```python
element_list=[2,2,4,4,5,2,3,6,5,6,2,4,8]
frequency_dic={}
for x in element_list:
     c=element_list.count(x)
     frequency_dic[x]=c
frequency_dic
```

###  4. Write a program to merge two dictionaries and resolve common keys by keeping the larger value

```python
dict1={"Name":"Ayush",
        "class":"B.Tech C.S.E"}
dict2={"address":"Dharuhera",
        "age":"21"}
merge_dict={}
for key1,key2 in zip(dict1,dict2):
    merge_dict[key1]=dict1[key1]
    merge_dict[key2]=dict2[key2]
merge_dict
```

### 5.Using a dictionary comprehension, create a dictionary from a list of integers where keys are numbers and values are their cubes, but only for odd numbers.

```python
integer_list=[1,2,3,4,5,6,7,8,9,0]
{elements:elements**3  for elements in integer_list if elements%2!=0}
```
